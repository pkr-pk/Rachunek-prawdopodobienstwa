###Minimum niezależnych zmiennych losowych
Jeżeli $X$ i $Y$ to niezależne zmienne losowe odpowiednio o dystrybuantach $F_X(x)$ i $F_Y(y)$ to wówczas zmienna losowa:

$Z = min(X, Y),$

ma rozkład o dystrybuancie

$F_Z(z) = 1 − (1 − F_X(z))(1 − F_Y (z)). \tag{1}$

Gdy zmienne $X$ i $Y$ mają jednakowy rozkład o dystrybuancie $F$, to:

$F_Z(z) = 1 − (1 − F(z))^2. \tag{2}$

Ogólnie, dla $n$ niezależnych zmiennych losowych $X_1, X_2,...,X_n$ o jednakowym rozkładzie o dystrybuancie $F$ zmienna losowa:

$Z = min(X_1,...,X_n),$

ma rokład o dystrybuancie:

$F_Z(z) = 1 − (1 − F(z))^n. \tag{3}$

###Maksimum niezależnych zmiennych losowych
Jeżeli $X$ i $Y$ to niezależne zmienne losowe odpowiednio o dystrybuantach $F_X(x)$ i $F_Y(y)$ to wówczas zmienna losowa:

$Z = max(X, Y),$

ma rozkład o dystrybuancie

$F_Z(z) = F_X(z)F_Y(z)). \tag{4}$

Gdy zmienne $X$ i $Y$ mają jednakowy rozkład o dystrybuancie $F$, to:

$F_Z(z) = (F(z))^2.\tag{5}$

Ogólnie, dla $n$ niezależnych zmiennych losowych $X_1, X_2,...,X_n$ o jednakowym rozkładzie o dystrybuancie $F$ zmienna losowa:

$Z = max(X_1,...,X_n),$

ma rokład o dystrybuancie:

$F_Z(z) =(F(z))^n.\tag{6}$

###Gęstość prawdopodobieństwa funkcji zmiennej losowej

Niech:

$u = u(x, y), \ \ \ v = v(x, y) \tag{7}$

będą funkcjami ciągłymi i różnowartościowymi o ciągłych pochodnych cząstkowych w pewnycm regularnym obszarze płaskim $D$. Załóżmy że (7), traktowane jako układ równań, ma dokładnie jedno rozwiązanie:

$x = x(u, v), \ \ \ y = y(u, v)$

o ciągłych pochodnych cząstkowych oraz, że jakobian:

$\frac{D(x,y)}{D(u, v)} = 
\begin{vmatrix}
    \frac{dx}{du} & \frac{dx}{dv} \\
    \frac{dy}{du} & \frac{dy}{dv} \\
\end{vmatrix} \ne 0 \ \ \ \text{dla $(u, v) \in \Delta$},$

gdzie $\Delta$ to obszar płaski płaszczyzny $OUV$, na który wzory (7) przekształcają obszar $D$.

Jeśli gęstość $f$ dwuwymiarowej zmiennej losowej $(X, Y)$ jest ciągła w $D$ poza skończoną liczbą linii, to gęstość $k$ dwuwymiarowej zmiennej losowej $(U, V)$ określonej wzorami:

$U = u(X, Y), \ \ \ V = v(X, Y)$

wyraża się wzorem:

$k(u, v) = \begin{cases}
f[x(u, v), y(u,v)]\left|\frac{D(x,y)}{D(u,v)}\right| & \text{dla $(u, v) \in \Delta$}, \\
0 & \text{dla $(u, v) \notin \Delta$}. \\
\end{cases} \tag{8}$

###Gęstość sumy, iloczynu, ilorazu zmiennych losowych
###Suma $Z = X + Y$

$f_Z(z) = \int\limits_{-\infty}^\infty f(x, z-x) dx, \tag{9}$

gdy zmienne są niezależne:

$f_Z(z) = \int\limits_{-\infty}^\infty f_X(x) f_Y(z-x) dx. \tag{10}$

###Iloczyn $Z = XY$

$f_Z(z) = \int\limits_{-\infty}^\infty \frac{1}{|x|}f(x, \frac{z}{x}) dx, \tag{11}$

gdy zmienne są niezależne:

$f_Z(z) = \int\limits_{-\infty}^\infty \frac{1}{|x|}f_X(x)f_Y(\frac{z}{x}) dx.\tag{12}$

###Iloraz $Z = \frac{X}{Y}$

$f_Z(z) = \int\limits_{-\infty}^\infty |y|f(zy, y) dy, \tag{13}$

gdy zmienne są niezależne:

$f_Z(z) = \int\limits_{-\infty}^\infty |y|f_X(zy) f_Y(y) dy. \tag{13}$

###Bezpośrednie wyznaczanie dystrybuanty funkcji zmiennych losowych

W wielu przypadkach zamiast gęstości wygodniej jest wyznaczyć dystrybuantę $F_Z$ zmiennej losowej $Z=g(X, Y)$ ze wzoru:

$F_Z(z) = \iint\limits_{D} f(x,y)dxdy, \tag{14}$

gdzie $D=\{(x, y): g(x,y)<z\}$
